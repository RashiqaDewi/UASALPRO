```python
import pandas as pd
import numpy as np

produk_data = {
    'Produk': ['produk 1', 'produk 2', 'produk 3', 'produk 4', 'produk 5', 'produk 6', 'produk 7'],
    'TipeKulit': ['Normal', 'Normal', 'Normal', 'Normal', 'Normal', 'Normal', 'Normal'],
    'Kategori': ['serum', 'serum', 'serum', 'serum', 'serum', 'serum', 'serum'],
    'Harga': [403, 310, 434, 465, 295, 434, 497],
    'Rating': [3.1, 3.9, 3.3, 3.4, 4.2, 3.8, 3.3]
}

df = pd.DataFrame(produk_data)

filtered_df = df[(df['Kategori'] == 'serum') & (df['TipeKulit'] == 'Normal') & (df['Rating'] > 0)].copy()
bobot = {'TipeKulit': 0.1, 'Kategori': 0.1, 'Harga': 0.6, 'Rating': 0.5}
normalized_df = filtered_df.copy()
print(normalized_df)
normalized_df['Harga'] = bobot['Harga'] * (normalized_df['Harga'] / normalized_df['Harga'].max())
normalized_df['Rating'] = bobot['Rating'] * (normalized_df['Rating'] / normalized_df['Rating'].max())
print("ini yang kedua", normalized_df)
ideal_positif = normalized_df.max()
ideal_negatif = normalized_df.min()
print("ini ideal positif", ideal_positif)
print("ini ideal negatif", ideal_negatif)
jarak_positif = (normalized_df[['Harga', 'Rating']] - ideal_positif[['Harga', 'Rating']]) ** 2
jarak_negatif = (normalized_df[['Harga', 'Rating']] - ideal_negatif[['Harga', 'Rating']]) ** 2
print( "ini jarak postif", jarak_positif)
print("ini jrak neg", jarak_negatif)
nilai_kedekatan = np.sqrt(jarak_negatif.sum(axis=1) / (jarak_positif + jarak_negatif).sum(axis=1))
print("ini nilai kedekatan", nilai_kedekatan)
filtered_df['NilaiKedekatan'] = nilai_kedekatan


hasil_rekomendasi = filtered_df.sort_values(by='NilaiKedekatan', ascending=True)

print(hasil_rekomendasi[['Produk', 'Harga', 'Rating']])
```
